```kotlin
%use dataframe
```

```kotlin
val df = DataFrame.read("data.csv")
df
```

```kotlin
val (md,mo, mk) = listOf(0.132, 0.035, 0.264)
val m = md + mo + mk

val g = 9.807
val r = 0.005 // радиус оси
val rn = 0.0006 // радиус нити
val r1 = 0.0425 // радиус диска
val r2 = 0.0525

m
```

```kotlin
val df1 = df.add("I"){ 
    val timeAvg = "t"<Double>() 
    val h0 = "h"<Double>()
    m * Math.pow(r + rn, 2.0) * ((g * timeAvg * timeAvg)/(2.0 * h0) - 1.0)
}.add("I_d"){
    (md * r1 * r1) / 2f
}.add("I_k"){
    (mk / 2f) * (r1 * r1 + r2 * r2)
}.add("I_teor"){
    1f/2f * (md * r1 * r1 + mk * (r1 * r1 + r2 * r2))
}

df1
```

```kotlin
%use lets-plot
```

```kotlin
letsPlot(df1.toMap()){
    x = "h"; y = "t"; z = "I";
} + ggsize(600, 500) + geomPoint()
```
